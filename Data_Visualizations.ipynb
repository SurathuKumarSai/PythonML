##Advanced Visualization - Seaborn

```
# importing packages 
import seaborn as sns 
import matplotlib.pyplot as plt 

# Apply the default theme
sns.set_theme()
```

```
## Load flights dataset
   
flights = sns.load_dataset("flights")

flights.head(15)
```

```
## Flights Data

(flights.shape)
```

####Filter Data

```

## Filter using loc

flights.loc[flights['month'] == 'May',:]

```

```

## Filter using query

flights.query(" month == 'May' ")
```

```

## Filter using loc

flights.loc[flights['passengers'] >= 500,:]
```

```
flights.query(" passengers >= 500 ")
```

```

## Filter using loc

flights.loc[(flights['passengers'] >= 500) & (flights['month'] == 'Aug') , ['year','passengers'] ]
```

```
flights.query(" passengers >= 500 & month == 'Aug' ")[['year','passengers']]
```

```
min_ppl = 500

flights.query(" passengers >= @min_ppl & month == 'Aug' ")
```

```
## Multiple values - loc

flights.loc[flights['month'].isin(['May','Jun']), : ]
```

```
## Multiple values - query

flights.query('month in ["May","Jun"] ') 
```

###Line Plot - Seaborn

```
may_flights = flights.loc[flights['month'] == 'May', : ]

may_flights
```

```
sns.lineplot(data=may_flights, x="year", y="passengers")
```

```
ax = sns.lineplot(data=may_flights, x="year", y="passengers")
ax.set_title('Passengers vs Year')
ax.set_ylabel('Passengers count')
```

```

may_june_flights = flights.loc[flights['month'].isin(['May','Jun']), : ]

may_june_flights_2 = flights.query(' month in ["May","Jun"] ')
```

```
sns.lineplot(data=may_june_flights, x="year", y="passengers", hue="month")
```

```
print(may_june_flights['month'].value_counts())
```

```
may_june_flights = flights.loc[flights['month'].isin(['May','Jun']), : ]

# Remove unused categories
may_june_flights['month'] = may_june_flights['month'].cat.remove_unused_categories()

print(may_june_flights['month'].value_counts())
```

```
sns.lineplot(data=may_june_flights, x="year", y="passengers", hue="month")
```

```

ax = sns.lineplot(data=may_june_flights, x="year", y="passengers", hue="month")

sns.move_legend(ax,loc = "best",bbox_to_anchor=(1, 1))
```

```


ax = sns.lineplot(data=may_june_flights, x="year", y="passengers", hue="month", marker="o")

sns.move_legend(ax,loc = "best",bbox_to_anchor=(1, 1))

```

```
## Task - Plot a line plot with dots for Jan, Feb, March months 

```

```

## Optional
## Add labels to points


ax = sns.lineplot(data=may_june_flights, x="year", y="passengers", hue="month", marker="o")


# Add value labels (annotations)
for line in ax.lines:
    for x, y in zip(line.get_xdata(), line.get_ydata()):
        ax.annotate(f'{int(y)}', xy=(x, y), xytext=(0, 5),  # Adjust text offset as needed
                    textcoords='offset points', ha='center', va='bottom', fontsize=8)
        
sns.move_legend(ax,loc = "best",bbox_to_anchor=(1, 1))
```

###Bar Plot - Seaborn

```
import seaborn as sns


# read a titanic.csv file from seaborn library
titan = sns.load_dataset('titanic')
 
titan
```

```
# who v/s fare barplot 

sns.barplot(x = 'who',y = 'fare',data = titan, ci=0)
```

```
from numpy import median

sns.barplot(x = 'who',y = 'fare',data = titan, estimator=median, ci=0)
```

```
# class v / s fare barplot with labels
ax = sns.barplot(x = 'who',y = 'fare',data = titan, ci = None)
ax.bar_label(ax.containers[0], fontsize=10);
```

```
# who v/s fare barplot in given order

sns.barplot(x = 'who',y = 'fare',data = titan, ci = None,
            order = ["woman", "man", "child"])
```

```
# who v/s fare barplot with class information

sns.barplot(x = 'who',y = 'fare',hue = 'class',data = titan, ci=0)
```

```
# who v/s fare barplot 
sns.barplot(x = 'who',y = 'fare',hue = 'class',data = titan,palette = "Blues", ci=0)
```

```
# who v/s fare barplot 
sns.barplot(x = 'who',y = 'fare',hue = 'class',data = titan,palette = "pastel", ci=0)
```

###Scatter Plot - Seaborn

```
tips = sns.load_dataset("tips")
tips.head()
```

```
sns.scatterplot(data=tips, x="total_bill", y="tip")
```

```
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="time")
```

```
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="time", style="time")
```

```
custom_markers = {"Lunch": "s", "Dinner": "X"}
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="time", style="time", markers=custom_markers)
```

```
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="day", style="time")
```

```
ax = sns.scatterplot(data=tips, x="total_bill", y="tip", hue="day", style="time",size="size")
sns.move_legend(ax,loc = "best",bbox_to_anchor=(1, 1))
```

```
ax = sns.scatterplot(data=tips, x="total_bill", y="tip", hue="day", style="time",size="size",sizes=(20,200))
sns.move_legend(ax,loc = "best",bbox_to_anchor=(1, 1))
```

```
## relational plots

sns.relplot(data=tips, x="total_bill", y="tip",col="time", hue="day", style="day",kind="scatter")
```
